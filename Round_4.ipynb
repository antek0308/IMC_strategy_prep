Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
from helpers import wrangle_backtest
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import numpy as np
```

Coconut and coconut_coupon data

```python
# COCONUT
coconut_df = trades_df[trades_df['symbol'] == 'COCONUT'].copy()
coconut_df['timestamp'] = pd.to_numeric(coconut_df['timestamp'], errors='coerce')
coconut_df = coconut_df.sort_values('timestamp')

print("=== COCONUT trades ===")
print(coconut_df)

# COCONUT_COUPON
coupon_df = trades_df[trades_df['symbol'] == 'COCONUT_COUPON'].copy()
coupon_df['timestamp'] = pd.to_numeric(coupon_df['timestamp'], errors='coerce')
coupon_df = coupon_df.sort_values('timestamp')

print("\n=== COCONUT_COUPON trades ===")
print(coupon_df)
```

### Plotting COCONUT and COCONUT_COUPON price

```python
# COCONUT
instrument = 'COCONUT'
instrument_df = trades_df[trades_df['symbol'] == instrument].copy()
instrument_df['timestamp'] = pd.to_numeric(instrument_df['timestamp'], errors='coerce')
instrument_df = instrument_df.sort_values('timestamp')

plt.figure(figsize=(10, 5))
plt.plot(instrument_df['timestamp'], instrument_df['price'], marker='o')
plt.title(f"Price of {instrument} over time (Round 4)")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()

# COCONUT_COUPON
instrument = 'COCONUT_COUPON'
instrument_df = trades_df[trades_df['symbol'] == instrument].copy()
instrument_df['timestamp'] = pd.to_numeric(instrument_df['timestamp'], errors='coerce')
instrument_df = instrument_df.sort_values('timestamp')

plt.figure(figsize=(10, 5))
plt.plot(instrument_df['timestamp'], instrument_df['price'], marker='o', color='orange')
plt.title(f"Price of {instrument} over time (Round 4)")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

COCONUT_COUPON shows greater relative price swings compared to COCONUT.

This makes sense if COUPON is an option, options are generally more sensitive to changes in the underlying.

Let's see the volatility

```python
plt.figure(figsize=(10, 5))

for instrument_loop in ["COCONUT", "COCONUT_COUPON"]:
    instrument_df = trades_df[trades_df['symbol'] == instrument_loop].copy()
    instrument_df['timestamp'] = pd.to_numeric(instrument_df['timestamp'], errors='coerce')
    instrument_df = instrument_df.sort_values('timestamp')
    log_returns = np.log(instrument_df["price"] / instrument_df["price"].shift())
    rolling_vol = log_returns.rolling(21).std().dropna() * np.sqrt(252)
    plt.plot(rolling_vol, label = f"{instrument_loop} rolling standard deviation (annualized)")
plt.legend()
```

The plot above shows the rolling monthly standard deviation (volatility) for both COCONUT and COCONUT_COUPON over time.

Both instruments experience periods of increased and decreased volatility, with noticeable spikes throughout the round.  
COCONUT_COUPON generally exhibits higher and more frequent volatility spikes compared to COCONUT.


### Correlation analysis

```python
merged = pd.merge_asof(
    coconut_df.sort_values("timestamp"),
    coupon_df.sort_values("timestamp"),
    on="timestamp",
    direction="nearest",
    suffixes=("_coconut", "_coupon")
)

correlation = merged["price_coconut"].corr(merged["price_coupon"])
print(f"Correlation: {correlation:.4f}")
```

Conclusions:

A high value close to +1 confirms that COCONUT_COUPON follows the base asset closely.

Linear regression (OLS)

COCONUT_COUPON ~ a * COCONUT + b

```python
X = merged["price_coconut"]
y = merged["price_coupon"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

Conclusions:

1. R-squared = 0.917
91.7% of the variance in COCONUT_COUPON is explained by COCONUT.
This is very high and confirms a strong linear relationship as COCONUT_COUPON behaves like a derivative of COCONUT.

2. Slope (coefficient for price_coconut) = 0.5033
Interpretation: for every 1-point change in the price of COCONUT, the COCONUT_COUPON price moves by ~0.50 points.
This resembles the delta of a call option, which often starts near 0.5.



```python
# Stationarity test for COCONUT and COCONUT_COUPON
for label, series in [("COCONUT", coconut_df["price"]), ("COCONUT_COUPON", coupon_df["price"])]:
    result = adfuller(series.dropna())
    print(f"{label} ADF Statistic: {result[0]:.4f}, p-value: {result[1]:.4f}")
```

Conclusions:

Stationarity Test Results

The Augmented Dickey-Fuller (ADF) test was used to check if the price series for COCONUT and COCONUT_COUPON are stationary.

Both p-values are much greater than 0.05, so we fail to reject the null hypothesis. This indicates that both price series are non-stationary, meaning their statistical properties (such as mean and variance) change over time.

### Linear regression (OLS)

COCONUT_COUPON ~ a * COCONUT + b

```python
import statsmodels.api as sm

X = merged["price_coconut"]
y = merged["price_coupon"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
predicted_coupon = model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(merged["price_coconut"], merged["price_coupon"], alpha=0.6, label="Observed data")
plt.plot(merged["price_coconut"], predicted_coupon, color="red", linewidth=2, label="Fitted line")

plt.xlabel("COCONUT Price")
plt.ylabel("COCONUT_COUPON Price")
plt.title("OLS Regression: COCONUT_COUPON vs. COCONUT (Round 4)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Key takeaways

91.7% of the variance in COCONUT_COUPON is explained by COCONUT.
This is very high and confirms a strong linear relationship — expected, as COCONUT_COUPON behaves like a derivative of COCONUT.

2. Slope (coefficient for price_coconut) = 0.5033
Interpretation: for every 1-point change in the price of COCONUT, the COCONUT_COUPON price moves by ~0.50 points.
This resembles the delta of a call option, which often starts near 0.5.
